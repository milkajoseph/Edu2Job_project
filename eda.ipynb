```python
def map_exp_cat(x):
    if pd.isna(x): return pd.NA
    s = str(x).strip().lower().rstrip(',')
    if s in ('entry','junior','fresh','fresher'): return 1.0
    if s in ('mid','mid-level','intermediate'): return 3.0
    if s in ('senior','lead','sr'): return 7.0
    m = re.search(r'(\d+(?:\.\d+)?)', s)
    return float(m.group(1)) if m else pd.NA
```

```python
df['experience_years'] = df['experience_level'].apply(map_exp_cat)
df['experience_years_filled'] = df['experience_years'].fillna(df['experience_years'].median())
df['experience_was_missing'] = df['experience_years'].isna()
df
```

```python
def course_qual(x):
    if pd.isna(x): return 'Unknown'
    s = str(x).lower()
    if 'bachelor' in s or 'b.' in s: return 'Bachelor'
    if 'master' in s or s.startswith('m'): return 'Master'
    if 'mba' in s: return 'MBA'
    if 'phd' in s: return 'PhD'
    if 'diploma' in s: return 'Diploma'
    return str(x).strip().title()

df['qualification_course'] = df['qualification'].apply(course_qual)
```

```python
def split_to_list(x):
    if pd.isna(x): return []
    s = str(x)
    for sep in [';', '|', '/', '\\n', '\n', '\t']:
        s = s.replace(sep, ',')
    s = re.sub(r'\s+and\s+', ',', s, flags=re.I)
    return [p.strip() for p in s.split(',') if p.strip()]

df['skills_list'] = df['skills'].apply(split_to_list)
df['num_skills'] = df['skills_list'].apply(len)
```
