```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib

all_mappings = []

for col in df.select_dtypes(include='object').columns:

    df[col + '_text'] = df[col].astype(str).str.strip().str.rstrip(',')


    le = LabelEncoder()
    df[col + '_encoded'] = le.fit_transform(df[col + '_text'])


    joblib.dump(le, f'{col}_encoder.pkl')


    mapping_df = pd.DataFrame({
        'Column': col,
        'Text': le.classes_,
        'Encoded': le.transform(le.classes_)
    })

    all_mappings.append(mapping_df)

final_mapping_df = pd.concat(all_mappings, ignore_index=True)

final_mapping_df.to_csv('all_categorical_mappings.csv', index=False)

print("All categorical mappings saved in 'all_categorical_mappings.csv'")

```
